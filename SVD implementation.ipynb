```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv("ratings_small.csv")
meta=pd.read_csv("movies_metadata.csv")
df.drop(['timestamp'],axis=1,inplace= True)
df.head()
```

```python
pivoted = df.pivot(index="userId", columns="movieId", values="rating")
pivoted
```

```python
df[df['userId']==662][0:10]
```

## ` Normalising the data to be around 0 `

```python
avrg=pivoted.mean(axis=1)
centerd_pivoted=pivoted.sub(avrg,axis=0)
centerd_pivoted
```

## ` filling null values with 0 `

```python

centerd_pivoted.fillna(0,inplace=True)
centerd_pivoted
```

```python
from scipy.sparse.linalg import svds
U,sigma,V_T=svds(centerd_pivoted.to_numpy())
sigma=np.diag(sigma)
```

```python
display(U)
display(sigma)
display(V_T)
```

```python
display(U.shape)
display(sigma.shape)
display(V_T.shape)
```

```python
ans=np.dot(np.dot(U,sigma),V_T)
```

## ` Reurning data to it's normal values `

```python
ans=ans+avrg.values.reshape(-1,1)
display(ans)
display(pivoted)
```

```python
calc_pred_ratings_df=pd.DataFrame(ans)
calc_pred_ratings_df
```
